# Linear Algebra in Python

```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Solve System of Equation

2x+3y = 8    
5x+4y = 20

```python
# coefficent matrix
A = np.array([[2,3],[5,4]])
A
```

# constant matrix
B = np.array([8,20])
B

```python
x = np.linalg.solve(A,B)
x
```

## Solving Complex Equation

\begin{align*}
(1 + 2i)x + (2 - i)y &= 4 + i \\
(3 - i)x + 4y &= 3 + 4i
\end{align*}

```python
# Coefficients matrix A with complex numbers
A_complex = np.array([[1+2j, 2-1j], [3-1j, 4]])
A_complex
```

```python
# Constants matrix B with complex numbers
B_complex = np.array([4+1j, 3+4j])
B_complex
```

```python
# Solving for X in the complex case
X_complex = np.linalg.solve(A_complex, B_complex)
X_complex
```

## Linear Transformation

\begin{bmatrix}
\cos(45^\circ) & -\sin(45^\circ) \\
\sin(45^\circ) & \cos(45^\circ)
\end{bmatrix}

```python
# create vector
v = np.array([5,3])
v
```

```python
# rotation matrix
thetha = np.radians(90)

rotation_matrix = np.array([[np.cos(thetha),-np.sin(thetha)],
                            [np.sin(thetha),np.cos(thetha)]])

v_rotated = np.dot(rotation_matrix,v)
v_rotated
```

```python
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Orignal')
plt.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated')
plt.xlim(-5,8)
plt.ylim(-5,8)
plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.legend()
plt.title("Vector Linear Transformation")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## Shear Transformation

```python
# Define the shear matrix for shearing along the x-axis with k = 1.5
shear_factor = 2.5
shear_matrix = np.array([[1, shear_factor], [0, 1]])

# Applying the shear transformation
v_sheared = np.dot(shear_matrix, v)
v_sheared
```

```python
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Orignal')
plt.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated')
plt.quiver(0, 0, v_sheared[0], v_sheared[1], angles='xy', scale_units='xy', scale=1, color='g', label='Sheared')
plt.xlim(-5,12)
plt.ylim(-5,12)
plt.axhline(y=0)
plt.axvline(x=0)
plt.grid()
plt.legend()
plt.title("Shear Transformation")
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# Adjusting the plot to include the effect of shearing on the grid lines
# Creating a grid of points
x = np.linspace(-5, 10, 10)
y = np.linspace(-5, 10, 10)
X, Y = np.meshgrid(x, y)

# Applying shear transformation to the grid points
X_sheared, Y_sheared = np.dot(shear_matrix, [X.ravel(), Y.ravel()])
X_sheared = X_sheared.reshape(X.shape)
Y_sheared = Y_sheared.reshape(Y.shape)

# Plotting the sheared grid lines
plt.figure()
plt.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r', label='Original Vector')
plt.quiver(0, 0, v_rotated[0], v_rotated[1], angles='xy', scale_units='xy', scale=1, color='b', label='Rotated Vector')
plt.quiver(0, 0, v_sheared[0], v_sheared[1], angles='xy', scale_units='xy', scale=1, color='g', label='Sheared Vector')

for i in range(len(x)):
    plt.plot(X_sheared[i, :], Y_sheared[i, :], 'k--', linewidth=0.5)
    plt.plot(X_sheared[:, i], Y_sheared[:, i], 'k--', linewidth=0.5)

plt.xlim(-5, 20)
plt.ylim(-5, 20)
plt.title('2D Vector Transformations with Sheared Grid')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
